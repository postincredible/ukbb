# HES secondary dataframe long to wide

icd9/10 code will be stored in a vector per eid

```python
%%time

import os
import pandas as pd
import numpy as np

# icd9s

os.chdir("/temp_project/HES")

hesin_icd9s_long=pd.read_csv('ukb_hesin_diag9.tsv',delimiter='\t',encoding='utf-8')

hesin_icd9s_w=hesin_icd9s_long.set_index(['eid','record_id','arr_index']).unstack(level=2)
hesin_icd9s_w.columns = hesin_icd9s_w.columns.map('{0[0]}_arr{0[1]}'.format)
hesin_icd9s_wide=hesin_icd9s_w.reset_index()

hesin_icd9s_wide['vec_icd9s']=hesin_icd9s_wide[hesin_icd9s_wide.columns[hesin_icd9s_wide.columns.get_loc('diag_icd9_arr0'):hesin_icd9s_wide.columns.get_loc('diag_icd9_nb_arr0')]].apply(lambda x: ','.join(x.dropna().astype(str)),axis=1)
hesin_icd9s_wide['vec_icd9s_nb']=hesin_icd9s_wide[hesin_icd9s_wide.columns[hesin_icd9s_wide.columns.get_loc('diag_icd9_nb_arr0'):hesin_icd9s_wide.columns.get_loc('vec_icd9s')]].apply(lambda x: ','.join(x.dropna().astype(str)),axis=1)

hesin_icd9s_wide.count()

hesin_icd9s_wide_vec=hesin_icd9s_wide[['eid','record_id','vec_icd9s','vec_icd9s_nb']]
hesin_icd9s_wide_vec.columns=['eid','record_id','diag_icd9_sec','diag_icd9_nb_sec']
hesin_icd9s_wide_vec.to_csv('/projects/ps-janssen3/dsci-pa/yhuan162/temp_project/HES/ukb_hesin_diag9_wide_vec.tsv', index=False, sep='\t')

```

```python
%%time

import os
import pandas as pd
import numpy as np


# icd10s

os.chdir("/temp_project/HES")


hesin_icd10s_long=pd.read_csv('ukb_hesin_diag10.tsv',delimiter='\t',encoding='utf-8')

hesin_icd10s_long['idx'] = hesin_icd10s_long.groupby(['eid','record_id']).cumcount()
hesin_icd10s_long2=hesin_icd10s_long.drop('arr_index',axis=1)
list(hesin_icd10s_long2.columns)

hesin_icd10s_long2.columns=['eid', 'record_id', 'diag_icd10', 'diag_icd10_nb', 'arr_index']

hesin_icd10s_long3=hesin_icd10s_long2.set_index(['eid','record_id','arr_index'])

hesin_icd10s_w=hesin_icd10s_long3.unstack(level=2)

hesin_icd10s_w.columns = hesin_icd10s_w.columns.map('{0[0]}_arr{0[1]}'.format)
hesin_icd10s_wide=hesin_icd10s_w.reset_index()

hesin_icd10s_wide['vec_icd10s']=hesin_icd10s_wide[hesin_icd10s_wide.columns[hesin_icd10s_wide.columns.get_loc('diag_icd10_arr0'):hesin_icd10s_wide.columns.get_loc('diag_icd10_nb_arr0')]].apply(lambda x: ','.join(x.dropna().astype(str)),axis=1)
hesin_icd10s_wide['vec_icd10s_nb']=hesin_icd10s_wide[hesin_icd10s_wide.columns[hesin_icd10s_wide.columns.get_loc('diag_icd10_nb_arr0'):hesin_icd10s_wide.columns.get_loc('vec_icd10s')]].apply(lambda x: ','.join(x.dropna().astype(str)),axis=1)

#hesin_icd10s_wide.count()

hesin_icd10s_wide_vec=hesin_icd10s_wide[['eid','record_id','vec_icd10s','vec_icd10s_nb']]
hesin_icd10s_wide_vec.columns=['eid','record_id','diag_icd10_sec','diag_icd10_nb_sec']
hesin_icd10s_wide_vec.to_csv('/temp_project/HES/ukb_hesin_diag10_wide_vec.tsv', index=False, sep='\t')
```
